```python
import pandas as pd
from sqlalchemy import create_engine,inspect
```

```python
### Top 10 netflix
csv_file = "Resources/all-weeks-countries.csv"
Top_10_df = pd.read_csv(csv_file)
Top_10_df.head()

## Transformation 

Top_10_df_refined = Top_10_df[["country_name","country_iso2","week","weekly_rank","show_title","cumulative_weeks_in_top_10"]].copy()
Top_10_df_refined = Top_10_df_refined.dropna()
```

```python
# Netflix title 
csv_file = "Resources/netflix_titles.csv"
Titles_df = pd.read_csv(csv_file)
Titles_df.head()

## Transformation 

Titles_df_refined = Titles_df[["type","title","director","date_added","release_year","rating","duration"]].copy()

Titles_df_refined
```

```python
#filter by dates we have >Sep 25,2021
Top_10_df_refined = Top_10_df_refined[Top_10_df_refined["week"] <= '2021-09-25']
Top_10_df_refined
Top_10_df_refined.rename(columns = {"show_title":'title'}, inplace = True)
Top_10_df_refined
```

```python
# replacing date to something workable in titles 
from datetime import datetime as dt


dates =[]

#droping all nan date added
Titles_df_refined = Titles_df_refined.dropna(subset= ["date_added"])


#itterate through rows and convert dates 
for index, row in Titles_df_refined.iterrows():
    s=str(row["date_added"])
    try:
        format = '%B %d, %Y'
        d=dt.strptime(s,format)
        date = (d.strftime('%Y-%m-%d'))
        dates.append(date)
    except:
        format = ' %B %d, %Y'
        d=dt.strptime(s,format)
        date = (d.strftime('%Y-%m-%d'))
        dates.append(date)
        


Titles_df_refined["date_added"] = dates
Titles_df_refined

```

```python
Titles_df_refined
```

```python
Titles_df_refined['director'] = Titles_df_refined['director'].fillna('Unknown')
Titles_df_refined
```

### Connect to database

```python
rds_connection_string = "postgres:P0stgres@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Create the inspector and connect it to the engine
inspector = inspect(engine)

# Collect the names of tables within the database
inspector.get_table_names()
```

```python
#Load the dataframes to SQL database
Titles_df_refined.to_sql(name='movies', con=engine, if_exists='append', index=False)
```

```python
Top_10_df_refined.to_sql(name='movies_rank', con=engine, if_exists='append', index=False)
```

```python
#Confirm data has been added by querying the movies table
pd.read_sql_query('select * from movies', con=engine).head()
```

```python
#Confirm data has been added by querying the movies_rank table
pd.read_sql_query('select * from movies_rank', con=engine).head()
```
